```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


def func(x):
    # return np.log(x) + 1
    return (1/5)*(x**2)
x = np.arange(0,8,0.001)
y = np.array(list(map(func, x)))
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
n = 115 # number of equal subdivisions
start, end = 2,6
h = (end-start)/n

x0 = []
y0 = []
area_sum = 0
initial_v = start
start+=h
while start<=end:
    x0.append(start)
    dot = func(start)
    y0.append(dot)
    area_sum+=(dot*h)
    start+=h

print('area under the cureve',area_sum) # this is integral

fig, ax = plt.subplots()
ax.plot(x,y, label='line')
ax.scatter(x0, y0, color='red', label="area dots")
ax.bar(x0, y0, width=h, color='purple', alpha=0.7)
ax.set_ylim(0, ymax=10)
ax.legend()
plt.show()
```

# Quicksot Algorith

```python
li = [-2,-3,4,10,12,3,4]
 
```

```python
def quicksort(l):
    if len(l)<=1:
        return l
    pivot = l[0]
    left = [i for i in l if i<pivot]
    right = [i for i in l if i>pivot]
    middle = [i for i in l if i==pivot]
    ll, rr = quicksort(left), quicksort(right)
    return ll+middle+rr
```

```python
res = quicksort(li)
res
```
